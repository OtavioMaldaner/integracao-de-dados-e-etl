# obt_copa_do_brasil_matches

Esse notebook é responsável por criar uma tabela com as informações descritivas e fatos de todas as partidas da Copa do Brasil.

## Configurações Iniciais

### Importando as Bibliotecas


```python
from pyspark.sql import functions as F
```

### Carregando os Dados

```python
df_brazilian_cup_matches = (
    spark.read.csv(
        "/Volumes/workspace/default/data/Brazilian_Cup_Matches.csv",
        header=True,
        inferSchema=True
    )
    .filter(
        (F.col("home_goal") != "NA")
        & (F.col("away_goal") != "NA")
    )
)
```

## Tratamento dos Dados

```python
df_final = (
    df_brazilian_cup_matches
    .withColumn(
        "championship",
        F.lit("Copa do Brasil")
    )
    .withColumns({
        "home_goal": F.col("home_goal").cast("int"),
        "away_goal": F.col("away_goal").cast("int"),
    })
    .withColumn(
        "composite_key",
        F.concat_ws(
            "-",
            F.col("home_team"),
            F.col("away_team"),
            F.col("season"),
            F.col("round")
        )
    )
    .orderBy(F.desc("datetime"))
)
```

## Salvamento dos Dados

```python
%sql
CREATE TABLE IF NOT EXISTS ifrs_soccer.gold.obt_copa_do_brasil_matches (
    round INT COMMENT "Rodada do campeonato",
    datetime TIMESTAMP COMMENT "Data e hora da partida",
    home_team STRING COMMENT "Nome do time da casa",
    away_team STRING COMMENT "Nome do time visitante",
    home_goal INT COMMENT "Gols do time da casa",
    away_goal INT COMMENT "Gols do time visitante",
    season INT COMMENT "Temporada (ano)",
    championship STRING COMMENT "Nome do campeonato",
    composite_key STRING PRIMARY KEY COMMENT "Chave composta única da partida formada por: home_team-away_team-season-round"
)
USING DELTA
TBLPROPERTIES ("quality"="gold", "layer"="data-engineering", "delta.enableTypeWidening"="true")
COMMENT "Tabela OBT com os resultados das partidas da Copa do Brasil.";
```

```python
(
    df_final
    .write.mode("overwrite")
    .option("mergeSchema", True)
    .saveAsTable("ifrs_soccer.gold.obt_copa_do_brasil_matches")
)
```
